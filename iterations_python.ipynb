````python
# Write a program to iterate from 1 to 10 numbers
for i in range(1,10):
    print(i)
````

````python
#write a program to iterate name
name="shivaji"
for c in name:
    print(c)
````

````python
# Write a program to print names using a for loop
names=["roopa","priya","sreeshanth","aaditya"]
for n in names:
    print("name of my friend is",n)
````

````python
# Write a program to print even numbers between 1 and 20 using a for loop
for i in range(1,21):
    if(i%2==0):
     print(i,end=",")
````

````python
# Write a program to print odd numbers between 50 and 100 using a for loop
for i in range(50,101):
    if(i%2!=0):
     print(i,end=",")
        
````

````python
# Write a program to find the vowels from a string using a for loop
name=str(input("please enter your name here:"))
print(name)
for char in name:
    if(char in ["a","e","i","o","u"]):
        print(char)
````

````python
# Write a program to find non-vowels from a string using a for loop
name=str(input("please enter your name here:"))
print(name)
for char in name:
    if(char not in ["a","e","i","o","u"]):
        print(char,end=",")
        
````

````python
# Write a program to find the sum of all numbers between 1 and 20 using a for loop
total = 0

# Loop through numbers from 1 to 20
for i in range(1, 21):
    total += i

# Display the result
print("The sum of numbers from 1 to 20 is:", total)
````

````python
# Write a program to find the sum of all odd numbers between 1 and 1000 using a for loop
# Initialize the sum
total = 0

# Loop through numbers from 1 to 1000
for i in range(1, 1001):
    if i % 2 != 0:   # Check if the number is odd
        total += i

# Display the result
print("The sum of all odd numbers between 1 and 1000 is:", total)
````

````python
result=1

# Loop through numbers from 400 to 500
for i in range(500, 601):
    if(i%2==0):  
        result *= i

# Display the result
print("The sum of all odd numbers between 500 and 600 is:", result)
````

````python
# Write a program to find numbers in between 1 and 20 using a while loop
sn=1
en=20
while (sn<=en):
    print(sn,end=",")
    sn=sn+1
    
````

````python
# Write a program to print odd numbers in between 100 and 200 using a while loop
sn=100
en=200
while (sn<=en):
    if (sn%2==1):
        print(sn,end=",")
    sn=sn+1
````
