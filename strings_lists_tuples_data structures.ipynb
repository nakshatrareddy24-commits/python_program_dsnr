```python
#strings
name="vikas"
info='welcome to tia'
data='''welcome to pp
python 
analysis
full-stack development
machine learning 
artificial intelligence
'''
print(name)
print(info)
print(data)
```

```python
# concatenation - adding two or more strings
a1="python"
a2="program"
print(a1+a2)
print(a1+" "+a2)
```

```python
# string repetition - repeating string
name="python program,"
print(name*10)
```

```python
#accessing characters from a string
# two ways   1. forward index   2. backward index (negative index)
#forward index - left to right (0,1,2.....n-1)
#backward index - right to left (-1,-2,-3,...,-n)
info = "welcome to tia"
print(info)
print(info[0])
print(info[-3])
print(info[11])
```

```python
#slicing - accessing group of characters from string - [ start_index:end_index-1]
info = "welcome to tia"
print(info)
print(info[0:7])
print(info[11:14])
print(info[:-3])
print(info[-3:])
```

```python
#reverse string
info = "priya"
print(info)
print(info[::-1])
```

```python
#methods or function in strings
print(dir(str))
```

```python
#help function return data type description 
print(help(str))
```

```python
#methods syntax - string.method()
info="python program"
print(info)
print(info.capitalize())
print(info.title())
```

# List in data structure
1. list in a collection of elements enclosed by []
2. It allows duplicate elements
3. each element separated by ,
4. list is mutable - editable

```python
# create list
names=["shiva","vikas","divya","keshav"]
ages=[25,26,27,28]
print(ages)
print(names)
print(type(names))
print(type(ages))
```

```python
#concatenation of list
a1=[10,20,30,40,50]
a2=[10.25,20.25,12.50,15.50]
a3=a1+a2
print(a3)
```

```python
#repeat list
info=["shiva","priya",10,20,30,10.25,12.25,20.25,15.50]
print(info)
print(info*5)
```

```python
#accessing elements from list
#index based
info=["shiva","divya",10,20,30,1000.25,12500.26]
print(info)
print(info[1])
print(info[-6])
```

```python
#slicing
info=["shiva","divya",10,20,30,1000.25,12500.26]
print(info)
print(info[2:5])
print(info[0:2])
print(info[5:7])
```

```python
#methods or functions supported by lists
print(dir(list))
```

```python
#methods with description
print(help(list))
```

```python
#append
names=["shiva","vikram","divya"]
names.append("chandu")
print(names)
```

```python
#insert()
names=["shiva","vikram","divya"]
names.insert(1,"vikas")
print(names)
```

```python
#entend()
names=["shiva","vikram","divya"]
names.extend(["chandu","vikas"])
print(names)
```

```python
#pop()
numbers=[10,20,30,40,50]
print(numbers)
numbers.pop(2)
print(numbers)
```

```python
#remove()
numbers=[10,20,30,40,50]
print(numbers)
numbers.remove(20)
print(numbers)
```

```python
#sorting technique sort() & reverse
numbers=[5,1,2,3,4,6]
print(numbers)
numbers.sort(reverse=False)
print(numbers)
```
